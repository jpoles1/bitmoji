```python
import urllib.request
import json
from numpy.random import randint
from io import BytesIO
from PIL import Image
import pandas as pd
```

```python
def display_image_from_url(img_url):
    img_data = urllib.request.urlopen(img_url).read()
    return Image.open(BytesIO(img_data))
def write_file(filename, contents):
    file = open(filename, "w")
    file.write(contents)
    file.close()
```

```python
json_url = "https://api.bitmoji.com/content/templates?app_name=bitmoji"
json_data = urllib.request.urlopen(json_url).read()
raw_data = json.loads(json_data.decode())
```

```python
jordo_id = "934d9900-892a-4515-8beb-202d6c42a6ee"
mia_id = "b84252ec-d02e-4a50-8682-0445ff64379c"
```

```python
raw_data.keys()
```

```python
num_solo_comics = len(raw_data["imoji"])
num_partner_comics = len(raw_data["friends"])
print(num_solo_comics, num_partner_comics)
```

## Solo Comics

```python
user_id = jordo_id
img_html_list = ['<img src="%s" width=100/>' % raw_data["imoji"][i]["src"] % user_id for i in range(num_solo_comics)]
write_file("jordo.html", "".join(img_html_list))
```

```python
user_id = mia_id
img_html_list = ['<img src="%s" width=100/>' % raw_data["imoji"][i]["src"] % user_id for i in range(num_solo_comics)]
write_file("mia.html", "".join(img_html_list))
```

## Solo Poses

```python
user_id = jordo_id
img_html_list = ['<img src="https://render.bitstrips.com//render/%s/%s-v1.png" width=100/>' % (raw_data["imoji"][i]["comic_id"], user_id) for i in range(num_solo_comics)]
write_file("jordo_poses.html", "".join(img_html_list))
```

```python
user_id = mia_id
img_html_list = ['<img src="https://render.bitstrips.com//render/%s/%s-v1.png" width=100/>' % (raw_data["imoji"][i]["comic_id"], user_id) for i in range(num_solo_comics)]
write_file("mia_poses.html", "".join(img_html_list))
```

## Partner Comics

```python
first_id = mia_id
second_id = jordo_id
img_html_list = ['<img src="%s" width=100/>' % raw_data["friends"][i]["src"] % (first_id, second_id) for i in range(num_partner_comics)]
write_file("mia_jordo.html", "".join(img_html_list))
```

```python
first_id = jordo_id
second_id = mia_id
img_html_list = ['<img src="%s" width=100/>' % raw_data["friends"][i]["src"] % (first_id, second_id) for i in range(num_partner_comics)]
write_file("jordo_mia.html", "".join(img_html_list))
```
